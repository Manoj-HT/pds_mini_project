```python
# Data clean up and required filter of years (2015, 2016, 2017) for all required datasets.

def Data_cleanup_Yr_Filter(df):
    years_to_filter = [ 2015, 2016, 2017 ] # So we can compare past year vs current (2015) and future year with a prediction (2017).
    df = df.rename(columns={'Date/Time':'DateTime'}) # Renaming the Date/Time to Datetime to remove specical characters.
    df.columns = df.columns.str.replace('~','_',regex=False) # Removing specical characters.
    try:
        df['DateTime'] = pd.to_datetime(df['DateTime'], format='%Y-%m-%d %H:%M') # Check if the DateTime column is in YYMMDD format
    except ValueError:
        df['DateTime'] = pd.to_datetime(df['DateTime'], format='%d-%m-%Y %H:%M') # If not Convert to YYMMDD format
    year_filtered_dataframe = df[df['DateTime'].dt.year.isin(years_to_filter)]  # Filter the data by Years 2015 to 2017
    Filtered_col=year_filtered_dataframe.columns[1:] # Select all columns after DataTime
    year_filtered_dataframe[Filtered_col]=year_filtered_dataframe[Filtered_col].applymap(lambda x: round(x,2)) # Apply Decimal limit to 2
    return(year_filtered_dataframe) # Print the final outcome
```
