```python
fd_pc = pd.merge(fd, PC, left_index=True, right_index=True)
fd_pc
```

```python
#Product of Hours and power will give Generation in KWh its divided by 1000 to convert into MWh
fd_pc['Turbine_Model_A_Gen']=(fd_pc.hours * fd_pc.Turbine_Model_A)/1000
fd_pc['Turbine_Model_B_Gen']=(fd_pc.hours * fd_pc.Turbine_Model_B)/1000
fd_pc['Turbine_Model_C_Gen']=(fd_pc.hours * fd_pc.Turbine_Model_C)/1000

fd_pc
```

```python
Gen_A=fd_pc.Turbine_Model_A_Gen.sum()
Gen_B=fd_pc.Turbine_Model_B_Gen.sum()
Gen_C=fd_pc.Turbine_Model_C_Gen.sum()
if Gen_A> Gen_B and Gen_A > Gen_C:
    print(f"The highest Generating turbine model is : Model_A = {Gen_A}")
elif Gen_B > Gen_A and Gen_B > Gen_C:
    print(f"The highest Generating turbine model is: Model_B = {Gen_B}")
else:
    print(f"The highest Generating turbine model is: Model_C = {Gen_C}")
```
